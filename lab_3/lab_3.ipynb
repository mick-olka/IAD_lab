```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
# read data
dat = pd.read_csv("./data/Life Expectancy Data.csv", header=0)
dat.head()
```

# Розвідувальний аналіз даних - EDA

```python
dat.shape, dat.info() # dataframe info and form
```

```python
dat.describe(include='object') # overall statistics
```

```python
dat.isnull().sum() # statistics of null features
```

# Removing null data fields

```python
country_list = dat['Country'].unique()
keys_list = dat.keys()
keys_list
```

```python
for country in country_list:
    dat.loc[dat['Country'] == country, keys_list] = dat.loc[dat['Country'] == country, keys_list].interpolate()
dat.dropna(inplace=True)
dat.isnull().sum()
```

```python
dat.info()
```

```python
pd.options.display.float_format = '{:.2f}'.format
dat.describe().T
```

# Графіки - histplots

```python
sns.set(font_scale=1.2)
data_nums = dat.select_dtypes(include='number')
col_nums = data_nums.columns.drop('Year')
```

```python
plt.figure(figsize=(20, 30))
plt.subplots_adjust(wspace=0.3, hspace=None)

i = 1
for col_name in col_nums:
    plt.subplot(7,3,i)
    sns.histplot(data = dat[col_name],  bins = 25, color="#5a5")
    sns.rugplot(data = dat[col_name], height=-0.02, clip_on=False, color="#5f5")
    i +=1
```

# Графіки - boxplots

```python
plt.figure(figsize=(20,30))
plt.subplots_adjust(wspace=1, hspace=None)

i = 1
for col_name in col_nums:
    plt.subplot(6,4,i)
    sns.boxplot(data=dat, y=col_name, color="#5a5")
    i +=1
```

# Кореляція

```python
plt.figure(figsize=(20,14))
sns.heatmap(dat.corr(), annot=True, cmap='crest')
plt.show()
```

```python
plt.figure(figsize=(20,14))
sns.heatmap(dat.corr(method="spearman"), annot=True, cmap='crest')
plt.show()
```

У графіку раховано демографічні змінні, структуру доходу та рівень смертності.
Видно, що кількість населення не впливає. 
Значний вплив на очікувану тривалість життя мають наступні показники: Income composition of resources (Human Development Index in terms of income composition of resources (index ranging from 0 to 1)), BMI (Average Body Mass Index of entire population), Number of years of Schooling(years), percentage expenditure (Expenditure on health as a percentage of Gross Domestic Product per capita(%)).
Зворотній вплив мають такі показники: HIV/AIDS (Deaths per 1 000 live births HIV/AIDS (0-4 years)), thinness (Prevalence of thinness among children and adolescents for Age 5 to 9 and 10 to 19 (% ))

```python
levpe = px.scatter(dat.groupby('Country').mean(), x='Life expectancy ', y='percentage expenditure', color='GDP', template='plotly', title='<b> Life expectancy vs Percentage expenditure', opacity=0.5)
# levpe.update_traces(marker={'size': 10})
px.scatter(dat, x='Life expectancy ', y='percentage expenditure', color='Country', template='plotly', title='<b> Life expectancy & Percentage expenditure')
levpe
```

На графіку видно, що для зібльшення тривалості життя 65+ країни повинні збільшувати фінансування системи охорони здоров'я.

```python
px.scatter(dat.sort_values(by='Year'), y='infant deaths', x='Life expectancy ', template='xgridoff', color='Country', opacity=0.7, title='<b>Life expectancy & Infant deaths of countries every year')
```

```python
px.scatter(dat.sort_values(by='Year'), y='Adult Mortality', x='Life expectancy ', template='plotly', color='Country', opacity=0.7, title='<b>Life expectancy & Adult mortality of countries every year')
```

Висновки: 
рівень смертності дорослих обернено пропорційно впливає на очікувану тривалість життя.

Коефіцієнт кореліцяї між тривалістю життя та BMI (body mass index) = 0.6, отже можна зробити висновки що спосіб життя та звички в їжі позитивно корелюють з тривалістю життя.
Рівень споживання алкоголю також позитивно корелює з тривалістю життя = 0,39.
А тривалість життя негативно корелює з відсотком дітей з пониженою вагою.

```python
px.scatter(dat, y=' BMI ', x='Life expectancy ', template='xgridoff', color='Country', opacity=0.7, title='<b>Life expectancy & BMI')
```
